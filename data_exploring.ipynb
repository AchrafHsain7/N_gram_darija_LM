```python
import unicodedata
from utils import DarijaBPETokenizer, NgramModel
```

# Loading the Data

```python
with open("./data/story_data_1.txt", encoding='utf-8') as f:
    corpus = f.readlines()

len(corpus)
```

```python
corpus
```

```python
corpus[0]
```

# Preprocessing

```python
# removing all new lines and double points 
corpus = [ x  for x in corpus if not ("[" in x and "]" in x)]
print(len(corpus))
corpus[0:4]
```

# Unique Characters

```python
corpus_chars = set()
for sent in corpus:
    for ch in sent:
        try:
            unicodedata.name(ch)
            corpus_chars.add(ch)
        except:
            continue
        
corpus_chars = list(corpus_chars)
```

```python
print(corpus_chars[:])
print(len(corpus_chars))
```

# BPE Tokenization

```python
tokenizer = DarijaBPETokenizer(corpus)
```

```python
len(tokenizer.train_vocab)
```

```python
tokenizer.bp_encode(max_vocab=1000)
tokenizer.train_vocab
```
